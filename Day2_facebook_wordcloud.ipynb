```python
from bs4 import BeautifulSoup

with open('your_posts_1.html',encoding="utf8") as page :
    soup = BeautifulSoup(page, 'html.parser')
```

```python
soup
```

```python
contents = soup.find_all('div', class_="_2pin")
[i.get_text() for i in contents]
```

```python
x = []
for i in contents:
    x.append(i.get_text())
```

```python
x
```

```python
y = " ".join(x) # joins all text
y = y.replace("Updated","")
y = y.replace("Feb ","")
y = y.replace("Click for video: ","")
y = y.replace("PM","")
y = y.replace("Jan ","")
y
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import random

wordcloud = WordCloud(background_color="white", max_words=2000,width=1024,height=720)
wordcloud.generate(y)
plt.figure(figsize=(20, 8))
plt.title('My First Facebook Wordcloud')
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show
```
